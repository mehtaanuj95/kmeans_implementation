# Elbow method

This method is used to find optimal number of clusters for kmeans-clustering.
The point after which the slope change becomes insignificant, then that point is termed as elbow point and that point is the optimal number of clusters.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.cluster import KMeans
```

```python
iris = datasets.load_iris()
data = iris.data
```

```python
# Visualize data
x = data[:,0]
y = data[:,1]
plt.scatter(x,y)
plt.show()
```

```python
num_of_clusters = range(1,9)
inertias = []

for k in num_of_clusters:
    model = KMeans(n_clusters=k)
    model.fit(data)
    inertias.append(model.inertia_)
    labels = model.predict(data)
    title = "Plot for {} clusters.".format(k)
    plt.title(title)
    plt.scatter(x, y, c=labels)
    plt.xlabel('sepal length (cm)')
    plt.ylabel('petal length (cm)')
    plt.show()
```

```python
# Plot elbow curve
plt.plot(num_of_clusters, inertias, '-o')

plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')

plt.show()
```
