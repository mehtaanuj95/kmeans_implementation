# K-means clustering using scikit-learn

```python
import matplotlib.pyplot as plt
from sklearn import datasets
import pandas as pd
import numpy as np
```

```python
# importing kmeans from sklearn's cluster
from sklearn.cluster import KMeans
iris = datasets.load_iris()
data = iris.data
```

```python
# Visualizing data
x = data[:,0]
y = data[:,1]
plt.scatter(x, y)

plt.xlabel('sepal length (cm)')
plt.ylabel('petal length (cm)')
plt.show()
```

```python
# Creating a model for 3 clusters
model = KMeans(n_clusters = 3)

# Fit the data to model
model.fit(data)

# Predict Labels
labels = model.predict(data)

# Print the labels
print(labels)
```

```python
# Visualizing the data with labels as colors
plt.scatter(x, y, c=labels)

plt.xlabel('sepal length (cm)')
plt.ylabel('petal length (cm)')
plt.show()
```

```python
# targets
target = iris.target

# Defining species
species = np.chararray(target.shape, itemsize=150)

# Categorizing labels
for i in range(len(data)):
  if target[i] == 0:
    species[i] = 'setosa'
  elif target[i] == 1:
    species[i] = 'versicolor'
  elif target[i] == 2: 
    species[i] = 'virginica'
    
df = pd.DataFrame({'labels': labels, 'species': species})

# Using cross-tabulation to print dtaframe
ct = pd.crosstab(df['labels'], df['species'])
print(ct)
```
